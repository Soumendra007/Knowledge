# AUC-ROC(Area Under the Curve-Receiver operating characteristic) curve:


 Imagine you're working on a machine learning project where you're building a model to distinguish between cats and dogs based on their images. The AUC-ROC curve is a way to understand how well your model is performing and making predictions.

AUC-ROC Curve Example:

Let's say you've trained your model and now you want to evaluate how good it is at distinguishing between cats and dogs. You have a dataset with labeled examples, some being actual cats and dogs.

    True Positive (TP): Your model correctly predicts an image as a cat.

    False Positive (FP): Your model incorrectly predicts an image as a cat when it's actually a dog.

    True Negative (TN): Your model correctly predicts an image as a dog.

    False Negative (FN): Your model incorrectly predicts an image as a dog when it's actually a cat.

        Now, let's create an ROC curve step by step:

You gather the predictions from your model for all the images in your dataset.

You sort these predictions from the most confident "cat" predictions to the least confident "cat" predictions.
As you do this, you start plotting points on your ROC curve. The x-axis represents the False Positive Rate (FPR), and the y-axis represents the True Positive Rate (TPR), which is also called sensitivity or recall.

FPR is calculated as FP / (FP + TN). It tells you how many times the model incorrectly predicted "cat" out of all the actual "dog" cases.

TPR is calculated as TP / (TP + FN). It tells you how many times the model correctly predicted "cat" out of all the actual "cat" cases.
By varying the threshold for your model's prediction (how confident it needs to be to classify something as a "cat"), you get different points on the curve. As the threshold changes, your FPR and TPR change, creating a curve that usually starts from (0,0) and moves toward (1,1).

Why AUC-ROC is Necessary:

The AUC-ROC curve helps you understand the trade-off between correctly identifying cats (TPR) and wrongly classifying dogs as cats (FPR) at various thresholds. It's especially useful because it provides a single number, the AUC (Area Under the Curve), that summarizes the overall performance of your model across all possible thresholds.

If your AUC is close to 1, it means your model is doing an excellent job of distinguishing between cats and dogs, regardless of the threshold you choose.

If your AUC is around 0.5, it suggests your model's predictions are close to random.

If your AUC is below 0.5, it means your model is performing worse than random guessing.

In essence, the AUC-ROC curve and the AUC value give you a concise way to compare different models and understand their performance in a way that's independent of the specific threshold you might choose for making predictions.

# Now I will perform LogisticRegression,GaussianNB,DecisionTree,SVC,KNeighborsClassifier with same dataset(breast_cancer:That I already use and spliting is also done for machine learning) using for loop and will print accuracy_score,confusion_matrix,classification_report one by one so lets see how can I do that: 

```python
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings("ignore")
```

```python
lg=LogisticRegression()
gnb=GaussianNB()
dtc=DecisionTreeClassifier()
svc=SVC()
knn=KNeighborsClassifier()
```

```python
model=[lg,gnb,dtc,svc,knn]

for m in model:
    m.fit(x_train,y_train)
    pred=m.predict(x_test)
    print("Accuracy score of ",m,"is:",accuracy_score(pred,y_test))
    print("Confusion matrix of",m,"is:",confusion_matrix(pred,y_test))
    print("Classification_report of",m,"is:",classification_report(pred,y_test))
```

The mode() function returns a pandas Series because there can be cases where multiple values have the same highest frequency (multiple modes). If you are interested in just the most common mode (the first one in case of multiple modes), you can use [0] to access that value from the Series.

# Encoder:

look at ds DataFrame now it has "City" and "Country" column that has string values.So when we will do data cleaning we can not use these two columns because we can not find z score for entire DataFrame or we can not find correlation for the entire ds DataFrame because of that two column.So what should I do shall we drop these column ,if we do that then we lost a huge portion of data.So to solve that we have two technique 1> One hot encoder and 2> Label encoder 

1> One hot encoder:- We use this method where column has only two string values it changes whose values to 0 or 1.

2> Label encoder:- We use this technique where column has more than 2 strings values it changes the strings to 0,1,2,3...if some strings repeats then it gives same value that it assign to that string value earlier.     


    Let see how can I do that:

```python
from sklearn.preprocessing import LabelEncoder

ln=LabelEncoder()
ds["City"]=ln.fit_transform(ds["City"])
```

```python
ds
```

Now you can see "City" column dont have any string values now.

```python
list1=["City","Country"]

for val in list1:
    ds[val]=ln.fit_transform(ds[val])
    
```
