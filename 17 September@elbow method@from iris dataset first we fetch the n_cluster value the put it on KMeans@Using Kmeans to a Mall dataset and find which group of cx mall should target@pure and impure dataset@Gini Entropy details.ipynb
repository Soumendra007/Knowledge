# The Elbow Method

To evaluate the performance of our k-means algorithm we can take a look at the inertia or objective function value. This is essentially the sum of squared distances our data points are away from their cluster centroid. By looking at different inertia values for different numbers of clusters (K)

Within Cluster Sum of Squares (WCSS)


                                                          WCSS is defined as the sum of the squared distance between each member of the cluster and its centroid.

![](wss.png)

For each k value, we will initialize k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster center.

Inertia_float Sum of squared distances of samples to their closest cluster center.


![](inertia_distence.png)

# Intra cluster distance

This is what inertia evaluates: It tells us how far the points within a cluster are. So, inertia actually calculates the sum of distances of all the points within a cluster from the centroid of that cluster

    We calculate this for all the clusters and the final inertial value is the sum of all these distances This distance within the clusters is known as intracluster distance so inertia gives us the sum of intracluster distances:


# Now I will load the iris dataset and import KMeans algo to perform unsupervise learning and will found what value I can put in n_clusters in KMeans algo by using elbow method.Lets see:

```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
iris=load_iris()
iris.data
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head(2)
```

```python
wcss=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km.fit(df)
    wcss.append(km.inertia_)
plt.plot(K,wcss,marker="x")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.title("Elbow method")
plt.show()
```

Here you can see for this df dataset the accurate elbow is 3 because for K value 3 is the right place where elbow starts.

# Now I know that I have to put n_clusters=3 in KMeans algo for this df dataset.Lets see how can I do that:

```python
model=KMeans(n_clusters=3)
y_model=model.fit_predict(df)
y_model
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
mall=pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset3/main/Mall_customers.csv")
mall.head(2)
```

```python
x=mall.iloc[:,3:]
x.head(2)
```

I want to get some clusters of customers according to their Annual Income (k$)and Spending Score (1-100) so I dont need "CustomerID","Genre","Age" these columns.

# Now I have to check what is the suitable cluster number I can put in n_cluster for KMeans algo for this "x" DataFrame.So I can check this out using elbow method lets see:

```python
#Elbow method
wcss=[]
for i in range(1,15):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)

plt.plot(range(1,15),wcss,marker="x")
plt.xlabel("cluster number")
plt.ylabel("Inertias")
plt.title("Elbow method")
plt.show()
    
```

So here you can see the best cluster value I can pick is 5 where elbow starts.

# Applying KMeans on my DataFrame "x"

```python
#Fitting KMeans to the dataset
km=KMeans(n_clusters=5,init="k-means++")
output=km.fit_predict(x)
output
```

Here init="k-means++" is default,it means it creats clusters group quickly.If you dont want to write init parameter no problem it will do the same work.

# Another detailed method of viewing all your datapoints mentionaed below:

```python
#ploting all the clusters and there centroied
#x axis: Anual income
#y axis: spending score
#s--> Marker size
plt.figure(figsize=(8,8))
plt.scatter(x[km.labels_==0]["Annual Income (k$)"],x[km.labels_==0]["Spending Score (1-100)"],s=100,c="green",label="Cluster 1")
plt.scatter(x[km.labels_==1]["Annual Income (k$)"],x[km.labels_==1]["Spending Score (1-100)"],s=100,c="red",label="Cluster 2")
plt.scatter(x[km.labels_==2]["Annual Income (k$)"],x[km.labels_==2]["Spending Score (1-100)"],s=100,c="yellow",label="Cluster 3")
plt.scatter(x[km.labels_==3]["Annual Income (k$)"],x[km.labels_==3]["Spending Score (1-100)"],s=100,c="blue",label="Cluster 4")
plt.scatter(x[km.labels_==4]["Annual Income (k$)"],x[km.labels_==4]["Spending Score (1-100)"],s=100,c="violet",label="Cluster 5")

#plot the centroids
#cluster_centers_ndarray of shape (n clusters, n_features) Coordinates of cluster centers.
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c="cyan",label="Centroid")

plt.title("Customer Group")

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
