# Classification:

You have to use classification when you deal with categorical data.There input variable can be categorical output variable also can categorical.Output can have multiple values or binary values.

There are different types of algorithms we can use for classification these are :-
    
    1> Naive Bayes Classifier
    
    2> Decision Tree Classifier
    
    3> K Nearest Neighbor Classifier (KNN)

    4> Support Vector Machine Classifier (SVM)
 
    

These are the major algorithm we use in classification.Every algorithm has its own story.

# 1> Naive Bayes Algorithm:-

In Naive Bayes algorithm first it creates a frequency table and based of that frequency table it creates a probability table. 

There are two types of Naive Bayes Algorithm.These are:-
    
    1> Multinomial NB(MNB):- We have to use Multinomial NB when our output column has more than two category outputs.Like if you remember we work a dataset named Iris dataset there are three out like Sitosa,Versicular,Verginica so the output column has 3 output category data so we have to use MNB here.
    
    2>Gaussian NB:-We have to use Gaussian NB when our output column has two category outputs only.Like if you remember we work on a dataset named Breast_cancer dataset that have only two output 0 or 1 so here we can use Gaussian NB to predict result.

# So now I will load a dataset named iris dataset and after loading I will not perform any EDA on it but its mandatory for you to perform EDA to clean,manipulation,outliers removal process before entering the machine learning stage.I am skip EDA part because I just want to perform Multinomial NB on this dataset.Let see how can I do that:

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings("ignore")
```

```python
irisds=load_iris()
irisds
```

```python
irisds.keys()
```

```python
irisds.data
```

```python
irisds.target
```

```python
irisds.feature_names
```

```python
df=pd.DataFrame(irisds.data,columns=irisds.feature_names)
df
```

```python
df["Target"]=irisds.target
df
```

# Spliting data for input and output variable:

```python
x=df.iloc[:,:-1]
x.head(2)
```

```python
y=df.iloc[:,-1]
y.tail(2)
```

# Spliting data for training for machine learning(MultinomialNB):

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Calling MultinomialNB and access it with and instance named mnb and all feed data to machine using fit() method:

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

# Testing machine using predict() method:

```python
pred=mnb.predict(x_test)
pred
```

# Checking accuracy_score,confusion_matrix,classification_report:

```python
print(accuracy_score(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

Here you can see 0 means setosa it has 19 value ad there is no error so classification report shows 1.00 means 100% data is correct.1 means versicolor it has 12 confirm count in test but machine have doubt for one value so accuracy is reduced for versicolor thats why classification shows .92 means 92% value is accurate for versicolor.2 means virginica it also  have 12 confirm count in test but machine have doubt for one value so accuracy is reduced for virginica to .92 that means 92% prediction is accurate .
